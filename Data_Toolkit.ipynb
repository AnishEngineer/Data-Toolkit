#2. 7A How does broadcasting work in NumPy?
Ans:- Broadcasting in NumPy allows operations on arrays with different shapes and sizes. Here's how it works:

Rules:
1. Matching dimensions: If arrays have different numbers of dimensions, the shape of the array with fewer dimensions is padded with ones on the left.
2. Size matching: If the size of the arrays in a particular dimension doesn't match, the size of the array with size 1 is "broadcasted" to match the size of the other array.

Example:

import numpy as np

a = np.array([1, 2, 3])  # shape: (3,)
b = np.array([4])  # shape: (1,)

result = a + b  # broadcasting happens here
print(result)  # Output: [5 6 7]


In this example, the array b with shape (1,) is broadcasted to match the shape of a, which is (3,). The resulting array has the same shape as a.

Benefits:
1. Vectorized operations: Broadcasting enables vectorized operations, making code more concise and efficient.
2. Flexibility: Broadcasting allows operations on arrays with different shapes and sizes.

#17.Explain the difference between apply() and map() in Pandas.
Ans:-Apply() vs Map() in Pandas

Apply():
1. Flexible function application: Applies a function to each row or column of a DataFrame.
2. Can handle complex operations: Can perform complex operations, including those that involve multiple columns.

Map():
1. Element-wise operation: Applies a function element-wise to a Series.
2. Typically used for simple operations: Often used for simple operations, such as replacing values or performing basic calculations.

Key differences:
1. Scope: apply() can operate on entire rows or columns, while map() operates on individual elements.
2. Function complexity: apply() can handle more complex functions, while map() is better suited for simple operations.

Choosing between Apply() and Map():
1. Use apply() for:
    - Complex operations involving multiple columns.
    - Operations that require access to entire rows or columns.
2. Use map() for:
    - Simple, element-wise operations.
    - Replacing values or performing basic calculations on individual elements.

Understanding the differences between apply() and map() helps you choose the most suitable method for your specific use case.

```python
#How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load example data (you can replace this with your own DataFrame)
df = sns.load_dataset("iris")  # Iris dataset has numeric columns: sepal_length, sepal_width, etc.

# 2. Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=['number']) # Select only numeric columns

# 3. Compute correlation matrix on numeric columns
corr = numeric_df.corr()

# 4. Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr,                      # correlation matrix
    annot=True,                # write correlation coefficients on the squares
    fmt=".2f",                 # numeric format, 2 decimal places
    cmap="coolwarm",           # color palette
    linewidths=0.5,            # lines between squares
    cbar_kws={"shrink": .8}    # colorbar sizing
)
plt.title("Correlation Matrix Heatmap")
plt.tight_layout()
plt.show()
```

```python
#Perform matrix multiplication using NumPy.
import numpy as np

# Define two matrices (2×3) and (3×2)
A = np.array([[1, 2, 3],
              [4, 5, 6]])
B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

# Method 1: Using the @ operator
C1 = A @ B

# Method 2: Using np.dot()
C2 = np.dot(A, B)

# Method 3: Using np.matmul()
C3 = np.matmul(A, B)

print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nA @ B =")
print(C1)

# Verify all methods give the same result
assert np.array_equal(C1, C2) and np.array_equal(C2, C3)
```
